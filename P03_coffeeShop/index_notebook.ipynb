### Objetivo del proyecto
El objetivo principal es desarrollar un cuadro de mando en Power BI para una cafeteria, que permita monitorizar, analizar y optimizar el rendimiento de ventas de los productos durante 6 meses.

### Alcance del proyecto
- Mostrar información de ventas y clientes de forma mensual, por dias y horas.
- Ofrecer una visualización clara de la evolución de ventas durante 6 meses.
- Proveer detalles específicos sobre productos y sus categorias con mejor y peor desempeño.
- Incluir indicadores de rendimiento clave (KPIs) relevantes para la gestión de ventas.

### Requerimientos funcionales
- Fuentes de Datos
- Visualizaciones y Elementos del Cuadro de Mando
- KPIs Generales
- Evolución de las Ventas
- Análisis de Productos
- Tabla de Productos Más Vendidos
- Número de Unidades Vendidas
- Filtros y Segmentación

### Plan de validacion
- Revisión de Datos
- Pruebas de Visualización
- Pruebas de Rendimiento
