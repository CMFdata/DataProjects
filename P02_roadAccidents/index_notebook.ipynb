# **Proyecto Traffic Accidents**
### Analisis de accidentes en carreteras federales de la región

image.png

### Objetivo del proyecto
Para este proyecto de expansión, necesitamos un estudio detallado sobre los registros de accidentes en carreteras federales.
Será necesario desarrollar un Dashboard que contenga información correspondiente a los años 2018, 2019 y 2020.

### Alcance del proyecto
- Total de Sucesos.
- Fallecidos y Heridos.
- Mapa de Geolocalizacion.
- Top 7 de Causas
- Accidentes con mas de 3 victimas.

### Requerimientos funcionales
- Fuentes de Datos
- Visualizaciones y Elementos del Cuadro de Mando
- KPIs Generales
- Total de Eventos por mes, año, dia de la semana, hora
- Análisis por Zona
- Tabla por zona y cantidad de victimas
- Filtro por año

### Plan de validacion
Revisión de Datos
- Pruebas de Visualización
- Pruebas de Rendimiento
